```
ham_word_counts_dict = dict(ham_word_counts.collect())
```

```
spam_word_counts_dict = dict(spam_word_counts.collect())
```

```
import math
```

```
def predict(sentence):
    words = sentence.split()
    spam_probs = []
    ham_probs = []
    results = []
    for word in words:
        # Spam
        if word in spam_word_counts_dict.keys():
            wordCount = spam_word_counts_dict[word]
        else:
            wordCount = 0
        pWordGivenSpam = (wordCount + 1)/(NSpam+V)
        spam_probs.append(math.log(pWordGivenSpam))
        # Ham
        if word in ham_word_counts_dict.keys():
            wordCount = ham_word_counts_dict[word]
        else:
            wordCount = 0
        pWordGivenHam = (wordCount + 1)/(NHam+V)
        ham_probs.append(math.log(pWordGivenHam))
    wordSpam = sum(spam_probs)+math.log(prob_spam)
    wordHam = sum(ham_probs)+math.log(prob_ham)
    if wordSpam>wordHam:
        return "spam"
    else:
        return "ham"
```

```
ans = test.map(lambda x: (predict(x[1]),x[0]))
```

```
correctPreds = ans.filter(lambda x: x[0]==x[1]).count()
```

```
incorrectPreds = ans.filter(lambda x: x[0]!=x[1]).count()
```

```
accuracy = correctPreds/(correctPreds+incorrectPreds)
```
