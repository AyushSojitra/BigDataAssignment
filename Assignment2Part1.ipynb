```
data = sc.textFile("/FileStore/shared_uploads/axs220055@utdallas.edu/assign2.txt")
```

```
data.collect()
```

```
userData = data.map(lambda x: (x.split('\t')[0],x.split('\t')[1]))
```

```
userData.collect()
```

```
userFriendPairs = userData.flatMap(lambda x: ((x[0],i) for i in x[1].split(',')))
```

```
userFriendPairs.collect()   
```

```
def findMutual(users):
    answer = []
    for user in users:
        userFriends = userData.lookup(user)
        userFriendsList = userFriends[0].split(',')
        userFriendsList.append(user)
        mutualFriends = userFriendPairs.map(lambda x: (x[0],1) if x[0] not in userFriendsList and x[1] in userFriendsList else (x[0],0)).reduceByKey(lambda x,y : x+y).sortBy(lambda x: -x[1]).filter(lambda x: x[1]>2).map(lambda x: x[0]).take(10)
        d = {"User": user, "FriendSuggestions": mutualFriends}
        answer.append(d)
    return spark.createDataFrame(answer)
```
